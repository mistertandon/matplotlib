```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
alpha = 0.7
phi_ext = np.pi

def flux_qubit_potential(phi_m, phi_p):
    return 2 + alpha - 2 * np.cos(phi_m) * np.cos(phi_p) - alpha * np.cos(phi_ext - 2*phi_p)
```

```python
phi_m = np.linspace(0, 2 * np.pi, 100)
phi_p = np.linspace(0, 2 * np.pi, 100)

X, Y = np.meshgrid(phi_m, phi_p)
Z = flux_qubit_potential(X, Y).T
```

```python
fig, ax = plt.subplots(nrows = 1, ncols = 1)
graph_i = ax.pcolor(X/(2*np.pi), Y/(2*np.pi), Z, cmap = matplotlib.cm.RdBu, vmin = abs(Z).min(), vmax = abs(Z).max())
fig.colorbar(graph_i, ax = ax)
```
