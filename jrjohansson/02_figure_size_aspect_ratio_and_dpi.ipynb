```python
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(1, 10, 10)
y = x ** 2
```

```python
fig, axes = plt.subplots(figsize = (8, 3), dpi = 100)
axes.plot(x, y)

axes.set_xlabel('x')
axes.set_ylabel('y')

axes.set_title('y = x ** 2 curve')
```
