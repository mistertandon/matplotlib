### Other 2d plot styles

```python
%config IPCompleter.greedy = True
```

```python
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
n = np.arange(5)
x = np.linspace(-0.75, 1, 100)
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 4, dpi=100)

axes[0].scatter(x, x ** 2)
axes[0].set_title('Scatter plot')

axes[1].step(n, n ** 2)
axes[1].set_title('Step plot')

axes[2].bar(n, n ** 2)
axes[2].set_title('Bar plot')

axes[3].fill_between(n, n ** 2, n ** 3, color = "blue")
axes[3].set_title('Fill between plot')

fig.tight_layout()
```

### Polar Graph

```python
fig_a = plt.figure(dpi = 100)

ax = fig_a.add_axes([0, 0, 0.6, 0.6], polar = True)
t = np.linspace(0, 2 * np.pi, 100)

ax.plot(t, t, color = "blue", linewidth = 2)
```
